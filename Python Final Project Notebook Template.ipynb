```python
data = {
    'Loan Status': ['Approved', 'Approved', 'Denied', 'Approved', 'Denied', 'Denied', 'Approved'],
    'Employment Status': ['Employed', 'Self-employed', 'Employed', 'Unemployed', 'Employed', 'Unemployed', 'Self-employed']
}
df = pd.DataFrame(data)
sns.countplot(data=df, x='Loan Status', hue='Employment Status')
plt.title('Loan Status by Employment Status')
plt.ylabel('Number of Applications')
plt.xlabel('Loan Status')
plt.legend(title='Employment Status')
plt.show()
```

```python
data = {
    'Rating (Alpha)': ['A', 'B', 'C', 'D', 'E', 'F'],
    'Rating (Numeric)': [5, 4, 3, 2, 1, 0],
    'Score': [85, 75, 65, 55, 45, 35]
}
df = pd.DataFrame(data)
plt.figure(figsize=(10, 6))
sns.barplot(x='Rating (Alpha)', y='Score', hue='Rating (Numeric)', data=df, palette="coolwarm")
plt.title('Chart of Rating (Alpha), Rating (Numeric), and Score')
plt.xlabel('Rating (Alpha)')
plt.ylabel('Score')
plt.show()
```

```python
data = {
    'Loan Status': ['Approved', 'Approved', 'Denied', 'Approved', 'Denied', 'Denied', 'Approved', 'Approved'],
    'Homeowner': ['Yes', 'No', 'Yes', 'Yes', 'No', 'No', 'Yes', 'No']
}

df = pd.DataFrame(data)

sns.countplot(data=df, x='Loan Status', hue='Homeowner')
plt.title('Loan Status by Homeowner Status')
plt.ylabel('Number of Applications')
plt.xlabel('Loan Status')
plt.legend(title='Homeowner')
plt.show()
```

```python
data = {
    'Estimated Loss': [1000, 2000, 1500, 3000, 2500, 1200, 1800, 2100],
    'Estimated Return': [5000, 7000, 6000, 10000, 8000, 5500, 7500, 7200]
}

df = pd.DataFrame(data)
sns.scatterplot(data=df, x='Estimated Loss', y='Estimated Return')
plt.title('Estimated Loss vs Estimated Return')
plt.xlabel('Estimated Loss')
plt.ylabel('Estimated Return')
plt.show() 
```

```python
data = {
    'Estimated Return': [5, 10, 15, 20, 25, 30, 35],
    'Estimated Loss': [2, 5, 1, 8, 7, 4, 6],
    'Investors': [100, 150, 200, 250, 300, 350, 400]
}
df = pd.DataFrame(data)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Estimated Return', y='Estimated Loss', size='Investors', hue='Investors', data=df, sizes=(100, 1000), palette="viridis", legend=False, alpha=0.7)
plt.title('Estimated Return, Loss, and Number of Investors')
plt.xlabel('Estimated Return (%)')
plt.ylabel('Estimated Loss (%)')
plt.show()
```

```python
data = {
    'Open Revolving Accounts': [2, 3, 4, 5, 3, 2, 4, 6],
    'Open Revolving Monthly Payment': [150, 200, 250, 300, 180, 160, 220, 280]
}
df = pd.DataFrame(data)
sns.scatterplot(data=df, x='Open Revolving Accounts', y='Open Revolving Monthly Payment')
plt.title('Open Revolving Accounts vs Open Revolving Monthly Payment')
plt.xlabel('Open Revolving Accounts')
plt.ylabel('Open Revolving Monthly Payment')
plt.show()
```

```python
data = {
    'Investors': [5, 10, 15, 20, 25, 30, 35, 40],
    'Percentage Funded': [50, 65, 70, 80, 85, 90, 95, 100]
}

df = pd.DataFrame(data)
sns.barplot(data=df, x='Investors', y='Percentage Funded',color='green')
plt.title('Percentage Funded by Number of Investors')
plt.ylabel('Number of Investors')
plt.xlabel('Percentage Funded'),
plt.ylim(0, 110) 
plt.show()
```

```python
data = {
    'Metric': ['lp_customer payments', 'lp_customer principal'],
    'Amount': [50000, 30000]  # Example values
}
df = pd.DataFrame(data)
sns.set(style="whitegrid")

sns.barplot(data=df, x='Metric', y='Amount')
plt.title('Comparison of LP Customer Payments and Principal')
plt.xlabel('Metric', color='blue')
plt.ylabel('Amount')
plt.ylim(0, max(df['Amount']) + 10000)  # Set y-axis limit for better visualization
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust the plot to ensure everything fits without overlapping
plt.show()
```

```python
years = list(range(2014, 2024))
records_last_10_years = [120, 130, 115, 140, 160, 180, 200, 220, 240, 260]
records_last_12_months = [220]
plt.figure(figsize=(10, 6))
plt.plot(years, records_last_10_years, label='Public Records Last 10 Years', marker='o')
plt.scatter(2023, records_last_12_months, color='red', label='Public Records Last 12 Months')
plt.axvline(x=2023, color='grey', linestyle='--', linewidth=1)
plt.xlabel('Year')
plt.ylabel('Number of Public Records')
plt.title('Relationship between Public Records Last 10 Years and Last 12 Months')
plt.legend()
plt.grid(True)
plt.show()
```
